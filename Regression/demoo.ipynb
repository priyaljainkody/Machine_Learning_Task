## 1 **Linear / Logistic Regression**

* **C (inverse regularization strength)**

  * Try values: `[0.01, 0.1, 1, 10, 100]`
  * Small C (0.01)  strong regularization  prevents overfitting.
  * Large C (100)  weak regularization  model fits more closely.

 Example: For text classification  `C = 1 or 10` often works best.

---

## 2 **Decision Tree**

* **max\_depth**: `[3, 5, 10, None]`
* **min\_samples\_split**: `[2, 5, 10, 20]`
* **min\_samples\_leaf**: `[1, 2, 5, 10]`

 Example:

* Small dataset  `max_depth=5`, `min_samples_split=5`.
* Large dataset  `max_depth=None`, `min_samples_leaf=2`.

---

## 3 **Random Forest**

* **n\_estimators**: `[100, 200, 500]` (more = better, but slower).
* **max\_depth**: `[5, 10, 20, None]`
* **min\_samples\_split**: `[2, 5, 10]`
* **max\_features**: `[sqrt, log2, None]`

 Example:

* `n_estimators=200`, `max_depth=10`, `max_features='sqrt'`.
* Gives stable accuracy with medium complexity.

---

## 4 **Gradient Boosting / XGBoost / LightGBM**

* **n\_estimators**: `[100, 300, 500, 1000]`
* **learning\_rate**: `[0.01, 0.05, 0.1, 0.2]`
* **max\_depth**: `[3, 5, 7, 10]`
* **subsample**: `[0.6, 0.8, 1.0]`
* **colsample\_bytree**: `[0.6, 0.8, 1.0]`

 Example:

* `learning_rate=0.1`, `n_estimators=300`, `max_depth=5`, `subsample=0.8`.
* Often a **good starting point** for tabular data.

---

## 5 **Support Vector Machine (SVM)**

* **C**: `[0.01, 0.1, 1, 10, 100]`
* **gamma** (RBF kernel): `[0.001, 0.01, 0.1, 1, 10]`
* **kernel**: `['linear', 'rbf']`

 Example:

* For small/clean data  `C=1`, `kernel='linear'`.
* For complex/nonlinear  `C=10`, `gamma=0.01`, `kernel='rbf'`.

---

## 6 **K-Nearest Neighbors (KNN)**

* **n\_neighbors (k)**: `[3, 5, 7, 11, 15]`
* **weights**: `[uniform, distance]`
* **metric**: `[euclidean, manhattan]`

 Example:

* Small dataset  `k=3`.
* Noisy dataset  `k=11 or 15`.
* If closer neighbors are more important  `weights='distance'`.

---

## 7 **Naive Bayes**

* **alpha (smoothing)**: `[0.1, 0.5, 1.0]`
* **variant**:

  * GaussianNB  for continuous features.
  * MultinomialNB  for text/word counts.
  * BernoulliNB  for binary features.

 Example:

* Text classification  `MultinomialNB(alpha=1.0)`.
* Binary features  `BernoulliNB(alpha=0.5)`.

 Example Workflow:
Say youre training a **Random Forest** on customer churn data:

* Start: `n_estimators=100`, `max_depth=None`.
* GridSearch: `n_estimators=[100, 200, 500]`, `max_depth=[5, 10, 20]`.
* Result: `n_estimators=200, max_depth=10`  better accuracy without overfitting.
