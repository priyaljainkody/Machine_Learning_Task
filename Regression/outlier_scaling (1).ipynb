```python
import pandas as pd
import numpy as np

def convert_discount(row):
    discount = row['Discount Applied']
    total_sales_amount = row['Total Sale Amount']

    if pd.isnull(discount) or pd.isnull(total_sales_amount):
        return np.nan

    discount_str = str(discount)

    if '$' in discount_str:
            value = float(discount_str.replace('$', ''))
            if value > total_sales_amount:
                return value/10
            else:
                percent = (value / total_sales_amount) * 100
                return percent

    elif '%' in discount_str:
            return float(discount_str.replace('%', ''))
    else:
        return 0.0

df['Discount Applied']= df.apply(convert_discount, axis=1)
df['Discount Applied']
```

```python
from scipy.stats import zscore
print('Before outlier removal : ',df.shape)
z = np.abs(zscore(df, axis = 1))
threshold = 3

dataset = df[(z < threshold).all(axis=1)]
print('After Z-Score approach : ',dataset.shape)
```

```python
Q1 = df.quantile(0.10)  # Q1
Q3 = df.quantile(0.90)  # Q3
IQR = Q3-Q1
df = df[~((df<(Q1-1.5*IQR)) | (df>(Q3+1.5*IQR)))]
print('After IQR approach : ',df.shape)
```
