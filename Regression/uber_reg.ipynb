```python
import matplotlib.pyplot as plt
 
# Group by hour and calculate mean fare
avg_fare_per_hour = df.groupby('hour')['fare_amount'].mean()
 
# Plot
plt.figure(figsize=(10,6))
plt.plot(avg_fare_per_hour.index, avg_fare_per_hour.values, marker='o')
plt.xticks(range(0,24))  # 0 to 23 hours
plt.xlabel("Hour of Day (0-23)")
plt.ylabel("Average Fare Amount ($)")
plt.title("Average Uber Fare by Hour of Day")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
 
# Convert pickup_datetime (UTC) to datetime
df['pickup_datetime'] = pd.to_datetime(df['pickup_datetime'], utc=True)
 
# Convert UTC to IST (India Standard Time)
df['pickup_datetime_ist'] = df['pickup_datetime'].dt.tz_convert('Asia/Kolkata')
 
# Extract IST hour and day_of_week
df['hour'] = df['pickup_datetime_ist'].dt.hour
df['day_of_week'] = df['pickup_datetime_ist'].dt.day_name()
 
# =======================
# Plot: Fare vs Hour
# =======================
plt.figure(figsize=(10,5))
df.groupby('hour')['fare_amount'].mean().plot(kind='line', marker='o')
plt.title("Average Fare Amount by Hour of Day (IST)")
plt.xlabel("Hour of Day (023)")
plt.ylabel("Average Fare Amount ($)")
plt.grid(True)
plt.show()
 
# =======================
# Plot: Fare vs Day of Week
# =======================
plt.figure(figsize=(10,5))
df.groupby('day_of_week')['fare_amount'].mean().reindex(
    ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
).plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Average Fare Amount by Day of Week (IST)")
plt.xlabel("Day of Week")
plt.ylabel("Average Fare Amount ($)")
plt.xticks(rotation=45)
plt.show()
```
