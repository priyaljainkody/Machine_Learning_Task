advantages of dbscan-> robus to outliers , no need to specify clusters , can find arbitary shaped clusters, only 2 hyperparameters to tune 

here we need three things -> core points, border points, noise points
here we have two hyperparameter for tuning -> epsilon, min_samples

```python
from sklearn.cluster import DBSCAN 
import numpy as np
import matplotlib.pyplot as plt 

X = np.array([[1,2], [2,2], [2,3], [8,7], [8,8], [25,80]])
```

```python
plt.scatter(X[:,0], X[:,1]) # x= [1,2,2,8,8,25] , y-coordinate = [2,2,3,7,8,80]
plt.show
```

```python
db = DBSCAN(eps=3, min_samples=2)
db.fit(X)
db.labels_ 

#here we have two clusters and one noise
```

```python
db2 = DBSCAN(eps = 3, min_samples = 3)
db2.fit(X)
db2.labels_
#here we have 1 cluter and two npise data points as we have 3 min_sample which says atleast 3 core point
```

```python
db1 = DBSCAN(eps=10, min_samples=2)
db1.fit(X)
db1.labels_

#here we have only one cluster and one noise as size of epsilon is big therefore it will make a large cirle in which more data points will come 
```

```python
db1 = DBSCAN(eps=10, min_samples=6)
db1.fit(X)
db1.labels_
# as there are less datapoints than min_samples therefore all will be consider as a noise 
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles 
from sklearn.cluster import DBSCAN
import numpy as np

X, _ = make_circles(n_samples = 500, factor = .5, noise = .03, random_state= 4)

dbscan = DBSCAN(eps= 0.1, min_samples=5)
clusters = dbscan.fit_predict(X)

plt.scatter(X[:,0], X[:,1], c=clusters, cmap = 'viridis', marker = 'o')
plt.title("DBSCAN clustering of concentric circles")
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.show()

# does not matter for dbscan which type of shape it is but kmeans can't handel 
```
